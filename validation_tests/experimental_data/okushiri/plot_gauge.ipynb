```python
# Main Imports
%pylab inline
import anuga
import numpy
```

```python
# Setup model variables
sww_filename = 'okushiri_auto_validation.sww'
gauge_locations = [[0.000, 1.696]] # Boundary gauge
gauge_locations += [[4.521, 1.196],  [4.521, 1.696],  [4.521, 2.196]] #Ch 5-7-9
gauge_names = ['Boundary', 'ch5', 'ch7', 'ch9']
```

```python
# Create a function to calculate time series data at gauges
wavefun = anuga.file_function(sww_filename,
                  quantities='stage',
                  interpolation_points=gauge_locations,
                  use_cache=False,
                  verbose=False)
```

```python
# Plot times series for specified gauge
gauge_id = 1; 
t = numpy.linspace(0.0,25.0,500);
y = map(lambda t: wavefun(t,gauge_id)[0], t);
plot(t,y); title(gauge_names[gauge_id]); xlabel('time (s)'); ylabel('stage (m)');
```
