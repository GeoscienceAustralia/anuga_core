```python
# Main Imports
%pylab inline
import anuga
import numpy
```

```python
# Setup model variables
sww_filename = 'okushiri_auto_validation.sww'
gauge_locations = [[0.000, 1.696]] # Boundary gauge
gauge_locations += [[4.521, 1.196],  [4.521, 1.696],  [4.521, 2.196]] #Ch 5-7-9
gauge_names = ['Boundary', 'ch5', 'ch7', 'ch9']
```

```python
# Create a function to calculate time series data at gauges
wavefun = anuga.file_function(sww_filename,
                  quantities='stage',
                  interpolation_points=gauge_locations,
                  use_cache=False,
                  verbose=False)
```

```python
# Read in validation data
fid = open('output_ch5-7-9.txt')
lines = fid.readlines()
fid.close()

final_time = 25.0
reference_time=[]
validation_data = {}
for j, key in enumerate(gauge_names[1:]):      # Omit boundary gauge
    validation_data[key] = [] 

        
for i, line in enumerate(lines[1:]):
    
    fields = line.split()
    time = float(fields[0])
    
    if time > final_time: break

    reference_time.append(time)    # Record reference time
    for j, key in enumerate(gauge_names[1:]):  # Omit boundary gauge
        value = float(fields[1:][j])           # Omit time 
        validation_data[key].append(value/100) # Convert cm2m

for j, key in enumerate(gauge_names[1:]):      # Omit boundary gauge
    validation_data[key] = numpy.array(validation_data[key]) 
```

```python
# Plot times series for specified gauge
gauge_id = 1; 
gn = gauge_names[gauge_id]
t = numpy.linspace(0.0,25.0,500);
y = map(lambda t: 400*wavefun(t,gauge_id)[0], t);
plot(t,y); title(gauge_names[gauge_id]); xlabel('time (s)'); ylabel('stage (m)');
plot(reference_time, 400*validation_data[gn]);
```
